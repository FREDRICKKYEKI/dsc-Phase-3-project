## 3.4 Checking for and removing multicollinearity (correlated predictors) 
In a classification problem, multicollinearity (high correlation between predictor variables) can still be a concern. While multicollinearity doesn't impact the accuracy or performance of classification models in the same way it does for regression models, it can affect the interpretability of the model and the stability of the coefficient estimates.

```python
# Visualize the correlation matrix
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

We now lookout for highly correlated variables, for this we set the threshhold at `0.7` to designate a **high** correlation.

```python
correlation_threshold = 0.7 # threshhold

highly_correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        # Get the correlation coefficient
        correlation = correlation_matrix.iloc[i, j]
        
        # Check if the correlation is above the threshold
        if abs(correlation) > correlation_threshold:
            # Get the variable names
            variable1 = correlation_matrix.columns[i]
            variable2 = correlation_matrix.columns[j]
            
            # Append the pair and correlation to the list
            highly_correlated_pairs.append((variable1, variable2, correlation))
```

```python
highly_correlated_pairs
```

From the output shown, the columns `'total day charge', 'total eve charge', 'total night charge', 'total intl charge'` introduce multicollinearity into our data-set, so as to ensure our model will not be affected by it we drop the columns for the following reasons:

**i. Multicollinearity**: High correlation between predictor variables can lead to multicollinearity, which makes it challenging to interpret the individual contributions of these variables to the model. When multicollinearity is present, it's difficult to disentangle the effects of correlated variables, and the coefficients may become unstable.

**ii. Interpretability**: In many cases, we want your model to be interpretable. Removing highly correlated variables simplifies the model and makes it easier to understand. This is particularly important if you need to communicate the results to non-technical stakeholders.

**iii. Model Efficiency**: Highly correlated variables may contain redundant information. Including all of them in the model doesn't necessarily improve predictive performance but can make the model computationally inefficient. Removing correlated variables can lead to faster training and prediction times.

**iv. Risk of Overfitting**: Correlated variables can increase the risk of overfitting. Overfit models are highly tuned to the training data but perform poorly on new, unseen data. Removing correlated variables can make your model more robust and less prone to overfitting.

```python
df = df.drop(columns=['total day charge', 'total eve charge', 'total night charge', 'total intl charge'])

# confirm if the multicollinearity is gone
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
