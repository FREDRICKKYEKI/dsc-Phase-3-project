## 3.4 Checking for and removing multicollinearity (correlated predictors) 
In a classification problem such as this, multicollinearity (high correlation between predictor variables) is a concern, and while multicollinearity doesn't impact the accuracy or performance of classification models in the same way it does for regression models, it can affect the interpretability of the model and the stability of the coefficient estimates.

In the heatmap we witnessed above ([here](#Bi-Variate-Analysis)), we saw that some variables are **highly** correlated. Let's try filter them out and see their relationships. 

For this, we set the threshhold at `0.7`, to designate a **high** correlation. and filter them.

```python
correlation_matrix = df.corr()

correlation_threshold = 0.7 # threshhold

highly_correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        # Get the correlation coefficient
        correlation = correlation_matrix.iloc[i, j]
        
        # Check if the correlation is above the threshold
        if abs(correlation) > correlation_threshold:
            # Get the variable names
            variable1 = correlation_matrix.columns[i]
            variable2 = correlation_matrix.columns[j]
            
            # Append the pair and correlation to the list
            highly_correlated_pairs.append((variable1, variable2, correlation))
```

```python
highly_correlated_pairs
```

From the output shown, the column pairs `'total day charge'` and `'total day minutes'`, `'total eve charge'` and ` 'total eve minutes'`, `'total night charge'` and `'total night minutes'` and finally `'total intl charge'` and `'total intl minutes'` introduce multicollinearity into our data-set.

So as to ensure our model will not be affected by it we drop one of each pair of columns for the following reasons:

**i. Multicollinearity**: High correlation between predictor variables can lead to multicollinearity, which makes it challenging to interpret the individual contributions of these variables to the model. When multicollinearity is present, it's difficult to disentangle the effects of correlated variables, and the coefficients may become unstable.

**ii. Interpretability**: In many cases, we want your model to be interpretable. Removing highly correlated variables simplifies the model and makes it easier to understand. This is particularly important if you need to communicate the results to non-technical stakeholders.

**iii. Model Efficiency**: Highly correlated variables may contain redundant information. Including all of them in the model doesn't necessarily improve predictive performance but can make the model computationally inefficient. Removing correlated variables can lead to faster training and prediction times.

**iv. Risk of Overfitting**: Correlated variables can increase the risk of overfitting. Overfit models are highly tuned to the training data but perform poorly on new, unseen data. Removing correlated variables can make your model more robust and less prone to overfitting.

From the highly correlated pairs, we note that one of each, contains usage (depending on the time, day, evening, night, or international), and the other contains the respective charges. We drop the column for charges and keep the minutes, because they  represent the actual usage. The charges can be derived using the respective tariff rates. 

So we drop "charge" columns.

```python
df = df.drop(columns=['total day charge', 'total eve charge', 'total night charge', 'total intl charge'])

# confirm if the multicollinearity is gone
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

Great! the heatmap now shows no case of multicollinearity of our data-set. We can prove it below.

```python
highly_correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        # Get the correlation coefficient
        correlation = correlation_matrix.iloc[i, j]
        
        # Check if the correlation is above the threshold
        if abs(correlation) > correlation_threshold:
            # Get the variable names
            variable1 = correlation_matrix.columns[i]
            variable2 = correlation_matrix.columns[j]
            
            # Append the pair and correlation to the list
            highly_correlated_pairs.append((variable1, variable2, correlation))

highly_correlated_pairs or "No more columns with a correlation of more than 0.7"
```

## 3.5 Converting categorical data to numeric format through one-hot encoding
From [this](#3.2-Data-Type-Conversions:) section of data type conversion and [categorical variables analysis](#Categorical-variables-analysis:) we were able to deduce that there are 5 categorical variables in our data-set (minus the target variable) with their respective value counts, namely:

But before we encode our data, there is an issue with our phone number column which is unique for each record. If we encode it, that means 3333 new records, which increases dimensionality and introduces noise into our dataset.

The remedy for this is to drop it, under the following reasons:

1. **High Cardinality:** 'Phone number' has too many unique values (3,333 in this case). Handling high cardinality categorical features can lead to memory and computational issues.

2. **Lack of Predictive Power:** 'Phone number' is unlikely to provide valuable information for predicting customer churn. Most machine learning models would not be able to extract meaningful patterns from it.

3. **Curse of Dimensionality:** Including a high-cardinality categorical variable can lead to the curse of dimensionality, making the dataset sparse and potentially affecting model performance.

4. **Data Privacy:** Depending on the context, the 'phone number' column may contain sensitive or personally identifiable information (PII). Handling such data requires careful consideration and may involve legal and privacy concerns.

5. **Practicality:** 'Phone number' may not be practically useful for analysis or modeling. Its inclusion may not align with the objectives of the project or the insights you want to gain from the data.

Considering these reasons, it's often more practical to drop or exclude the 'phone number' column from your analysis or modeling efforts.

```python
# dropping phone number
df = df.drop(columns=['phone number'])
```

```python
# confirm if it has been successfully dropped
print((df.shape))
df.head()
```

The dataset is now ready for encoding! We now filter out the categorical variables to encode them, so as to later concat them back to the numerical variables before the modelling process begins.  

```python
# These are the updated numerical and categorical columns
categorical_vars = list(df.select_dtypes(include=['object']).columns )
numerical_vars =  list(df.select_dtypes(include=['int64', 'float64']).columns )
categorical_vars, numerical_vars
```

The following code encodes the categorical variables, while dropping the first of each, to avoid any multicollinearity brought about by the encoded variables.

```python
# one hot encoding before modelling
df = pd.get_dummies(df, columns=categorical_vars, drop_first=True)
```
